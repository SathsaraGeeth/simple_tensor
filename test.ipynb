```python
from tensor import Tensor
import ctypes


sz = 2**14
```

```python
buf = (ctypes.c_float * (sz*sz))(*([1.0] * (sz*sz)))
t1 = Tensor([sz, sz], dtype='REAL32', buffer=buf, view=True)
t2 = t1.deep_copy()
```

```python
import time

start = time.time()

t3 = t1 + t2


end = time.time()
print(end - start)


# print(t3)
```

```python
import torch
import time

# Create tensors
t1 = torch.ones((sz, sz), dtype=torch.float32)
t2 = t1.clone() 

# Timing the addition
start = time.time()

t3 = t1 + t2
end = time.time()

print("Elapsed time:", end - start)

print(t3)
```
